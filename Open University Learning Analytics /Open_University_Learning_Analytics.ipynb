# **Overview**
Open University Learning Analytics Dataset (OULAD) contains data about courses, students and their interactions with Virtual Learning  Environment (VLE) for seven selected courses (called modules). Presentations of courses start in February and October - they are marked by "B" and "J"  respectively. The dataset consists of tables connected using unique identifiers. Dataset is stored in several csv files. Visit https://analyse.kmi.open.ac.uk/open_dataset for more information on the dataset.


This notebook explores the main drivers in outcomes for the different modules and presentations across demographics. The goal is to understand how student behavior may influence their performance and identify opportunities to improve the overall experience.

### **Recommendations**

**1. Encourage Community Learning**

Using VLE more frequently doesn't necessarily mean students will score better or have a higher chance of passing/distinction. VLE activity types play a crucial role in how students perform. In particular, data has shown that when students use forumng as their primary activity type, they average higher scores. 

Assuming forumng is related to online forums/communities, this tells us students perform better when they work together. We should create more community learning activities. Then measure the impact by comparing student outcomes and assessment scores to validate if increased community learning improves results.


**2. Changes Course Structure**

An opportunity to improve student outcomes is to develop course material that mirrors the structure of modules AAA or GGG. Though these modules have fewer students, they have lower withdrawals, higher passing rates, and more students using VLEs, which indicates students are engaged and thriving. 

Similarly, J (October) courses outperform B (February) with more sign-ups and better student outcomes. This is a sign that seasonality between start dates impacts student behaviors. We should update course structures to be similar to AAA, GGG, and J (October) start dates. This will enable us to see if course styles impact outcomes.


**3. Develop Additional Support and Resources**

We have seen that there are imbalances between student's outcomes based on their age, region, previous education level across all modules. It is a concern when students with lower previous education are withdrawing and failing at a higher rate. To ensure course material is suited for all levels,  we should run surveys or focus groups to understanding what is driving them to leave and how material can be improved.

# **Formatting Data**

Before diving deeper into the analysis, I want to add some key **`date`** columns to each CSV. Adding the dates will help understand time-series trends across the datasets. The process will consist of incorporating the number of days a specific data value has elapsed. 
* For example, if a user registered `-120 days` before the presentation start date, I will subtract this value and created a datestamp column.

This dataset states the **`code_presentation`** value consists of the **`year`** and **`B`** for the presentation starting in **February** and **`J`** for the presentation starting in **October**. I will assume and truncate the dates to the **1st day** of the respective month. 
* For example, if **`code_presentation = 2013J`** then I will assume the start date is **October 1, 2013**

### **Data Uploads**
I uploaded the CSV files to **Google BigQuery**, and I will be running my SQL queries in that environment and pipe any queries into this notebook for analysis and visualization.

For formatting, I cleaned and added all the date related values mentioned above in **Google BigQuery** and dropped and replaced the original CSV files to create new tables with the relevant data. This will ensure better readability and fast extraction of the date values throughout the analysis. Below are all the formatting queries I ran locally in **BigQuery**. 

### **New Tables**
**1. courses.csv** → adding start and end date
```
SELECT  
  *,
  case when code_presentation = '2013B' then cast('2013-2-1' as date)
       when code_presentation = '2013J' then cast('2013-10-1' as date)
       when code_presentation = '2014B' then cast('2014-2-1' as date)

       when code_presentation = '2014B' then date_add(cast('2014-2-1' as date), INTERVAL module_presentation_length DAY)
       when code_presentation = '2014J' then date_add(cast('2014-10-1' as date), INTERVAL module_presentation_length DAY) end as end_date
FROM `datasets-292703.adobe.courses` 
order by 1,4
```

**2. assessments.csv** → adding a due date
```
with days_formatting AS (
  SELECT  
    a.*,
    start_date,
    end_date,
    case when assessment_type = 'Exam' and date = '?' then -7 else cast(date as int64) end days
  FROM `datasets-292703.adobe.assessments` as a
  JOIN `datasets-292703.adobe.courses`  as c
    on a.code_module = c.code_module and a.code_presentation = c.code_presentation
  order by 1,3)

select
  code_module,
  code_presentation,
  id_assessment,
  assessment_type,
  days,
  weight,
  case when assessment_type = 'Exam' then date_add(end_date, INTERVAL days DAY)
           else date_add(start_date, INTERVAL days DAY) end as due_date
from days_formatting 
order by 1,3  
```

**3. vle.csv**→ adding a week start and end date
```
with vle_formatting AS (
  SELECT  
    id_site,
    v.code_module,
    v.code_presentation,
    activity_type	,
    cast((case when week_from = '?' then null else week_from end) as int64) week_from,
    cast((case when week_to = '?' then null else week_to end) as int64) week_to,
    start_date,
    end_date,
  FROM `datasets-292703.adobe.vle` as v
  JOIN `datasets-292703.adobe.courses`  as c

  )

select
  id_site,
  code_module,
  code_presentation,
  activity_type,
  week_from,

  case when week_from is null then start_date 
       else date_add(start_date, INTERVAL week_from week) end as week_start,
  case when week_to is null then end_date

```

## **1. Performance Funnel**

Looking at how the different modules went through the funnel in terms of "Performance," we can define the five steps as follows.

  1. **Signed Up** - The total number of students signed up
  2. **Used VLE** - The total number of students who used VLE at least once.
  3. **Submitted Assessment** - The total number of students who had submitted at least one assessment.
  4. **Continued** - The total number of students who were still registered (i.e., excluding the students who unregistered/withdrew
  5. **Passed + Distinction** - The total number of students that either passed or had distinction.


The methodology behind these steps' order is because across 6 out of the 7 modules, has the number of students drops gradually in volume as we move down the steps. There is one exception for module `GGG`, which has a higher volume for a lower step, which will be discussed shortly.

This funnel performance aims to understand which modules record higher results and there may have larger impacts on students' grades. This high-level view groups all students across both `B` and `J` presentations and year. Using this general overview will guide potential direction into further exploration.

Below we can view the performance funnel chart across each module.

### **Performance Funnel by Module**

#### **Insights**

````
query = """
with 
  
  -- NUMBER OF STUDENTS WHO SIGNED UP AND THE NUMBER WHO DIDN'T DROP OUT
  registrations AS (
  SELECT
    code_module,
    code_presentation,
    count(distinct id_student) as total_registered,
    count(distinct id_student) - count(distinct case when unregistration_date is not null then id_student end) as continued,
  FROM
    `datasets-292703.adobe.studentRegistration`
  GROUP BY 1,2)
  
  -- NUMBER OF STUDENTS WHO USED VLE
, vle_used AS (  
  SELECT
    code_module,
    code_presentation,
    count(distinct id_student) as used_vle
  FROM
    `datasets-292703.adobe.studentVle`
  GROUP BY 1,2)
  
  -- NUMBER OF STUDENTS WHO SUBMITTED AN ASSESSMENT
, assessments AS ( 
  SELECT
    a.code_module,
    a.code_presentation,
    count(distinct id_student) as submitted_assessment
  FROM
    `datasets-292703.adobe.studentAssessment`
  JOIN `datasets-292703.adobe.assessments` as a
    USING (id_assessment)
  JOIN `datasets-292703.adobe.courses` as c
    ON c.code_module = a.code_module AND c.code_presentation = a.code_presentation
  GROUP BY 1,2)
  
  -- FINAL RESULTS FROM THE MODULE/PRESENTATIONS
, results AS (  
  SELECT
    code_module,
    code_presentation,
    count (distinct case when final_result in ('Pass', 'Distinction') then id_student end) as passing_and_distinction,
    count (distinct case when final_result in ('Distinction') then id_student end) as distinction,
  FROM `datasets-292703.adobe.studentInfo`
  group by 1,2
  order by 1,2)
  
  SELECT
    r.code_module,
    r.code_presentation,
    total_registered,
    used_vle,
    submitted_assessment,
    continued,
    passing_and_distinction,
    distinction,
  FROM registrations as r
  JOIN vle_used as v
    ON r.code_module = v.code_module AND r.code_presentation = v.code_presentation
  JOIN assessments as a
    ON r.code_module = a.code_module AND r.code_presentation = a.code_presentation
  JOIN results as rs
    ON r.code_module = rs.code_module AND r.code_presentation = rs.code_presentation
  ORDER BY 1,2
"""

# saving above query as dataframe
performance_funnel = client.query(query).to_dataframe()
````

### **Scores by Assessment Type**

````
query = """
with 
  
  modules as (
  SELECT
    code_module,
    code_presentation,
    assessment_type,
    id_assessment
  FROM
    `datasets-292703.adobe.assessments`
  ORDER BY 1,2,3)

, median_score as (
  SELECT
    id_assessment,
    percentile_disc(score, 0.5) over(partition by id_assessment) as score
  FROM `datasets-292703.adobe.studentAssessment` 
  where score != '?')
  
, assessments as (
  SELECT
    id_assessment,
    count(*) as total_assessment_submitted,
    sum(is_banked) as transferred,
    round((sum(is_banked)/count(*)*100),2) as transferred_rate,
    avg(submission_days) as avg_submission_days
  FROM `datasets-292703.adobe.studentAssessment` as a
  GROUP BY 1
  oRDER BY 1)
  
, score_clean as (
  SELECT
    id_assessment,
    cast(case when a.score = '?' then null else a.score end as numeric) score
  FROM `datasets-292703.adobe.studentAssessment` as a)

, failed_scores as (
  select
    id_assessment,
    count(case when score < 40 then id_assessment end) as failed
  from score_clean
  group by 1
  order by 1)

  SELECT
    distinct m.*,
    a.* except(id_assessment),
    ms.score as median_score,
    failed,
    round((failed/total_assessment_submitted)*100,2) as failed_rate
  FROM modules as m 
  LEFT JOIN assessments as a
  USING (id_assessment)
  LEFT JOIN median_score as ms
  USING (id_assessment)
  LEFT JOIN failed_scores as f
  USING (id_assessment)
  ORDER BY 1,2,3
"""

assessment_scores = client.query(query).to_dataframe()
````

````
# formatting type
assessment_scores['median_score'] = assessment_scores['median_score'].astype('float')

# grouping by module and assessment type
temp = assessment_scores.groupby(by=['code_module', 'code_presentation', 'assessment_type'])\
                                    [['total_assessment_submitted', 'transferred', 'avg_submission_days', 'median_score', 'failed']]\
                        .agg({'total_assessment_submitted' : 'sum',
                              'transferred': 'sum',
                              'avg_submission_days': 'mean',
                              'median_score' : 'median',
                              'failed': 'median'})\
                        .round(1)\
                        .reset_index()

temp['transferred_%'] = round((temp['transferred']/temp['total_assessment_submitted'])*100,2)
temp['failed_%'] = round((temp['failed']/temp['total_assessment_submitted'])*100,2)
````

### **Average Score for Top Used Activity by Individual Student**
One strategy to understand the VLE activity type's value is to identify the number one used VLE per student and then check their average score. For example, if student A uses **subpage** vle type more than student B, who uses **forumng**, which student gets a better average score.

The histograms below show these insights, displaying the percentage distribution of students' average scores based on their **Number 1** used VLE activity type.

Across the top 6 most used activity types, the distributions are slightly left-skewed. **forumng** has a narrower distribution with a higher portion of students scoring between 80-90, which indicates the success of this activity type.
  * Students within this range represent +3% of the population for each score interval.
  * After identifying the number one used VLE activity type per student, we can see that forumng registered the highest median score, 79.

We see similar behavior with **quiz** except the range is lower, with the total student population represented by >3% for each score between 77-83.

````
query = """
WITH
  t AS (
  SELECT
    id_student,
    s.code_module,
    s.code_presentation,
    activity_type,
    SUM(sum_click) AS total_clicks,
    COUNT(*) AS vlew_cnt,
  FROM `datasets-292703.adobe.studentVle` s
  JOIN `datasets-292703.adobe.vle` v
    USING (id_site)
  GROUP BY 1, 2, 3,4
  ORDER BY 1,2, 4 DESC)
  
, ranks AS (  
  SELECT
    t.*,
    RANK() OVER(PARTITION BY id_student, code_module, code_presentation ORDER BY vlew_cnt DESC) AS vle_ranking,
    RANK() OVER(PARTITION BY id_student, code_module, code_presentation ORDER BY total_clicks DESC) as click_ranking
  FROM
    t)
    
, students as (    
  SELECT
  *
  FROM ranks
  where vle_ranking = 1
  ORDER BY 1,2,3, 5 desc)
  
, score_clean as (
  SELECT
    id_assessment,
    id_student,
    cast(case when a.score = '?' then null else a.score end as numeric) score
  FROM `datasets-292703.adobe.studentAssessment` as a)
  
, scores as (
  select
    id_student,
    round(avg(score)) as avg_score
  from score_clean
  group by 1)
  
  select
    s.id_student,
    code_module,
    code_presentation,
    activity_type,
    total_clicks,
    vlew_cnt,
    avg_score
  from students as s
  join scores  as v
  On s.id_student = v.id_student 
  order by 1,2,3,4
  """

top_rank = client.query(query).to_dataframe()
````

### **Correlation between Assessments, VLE Acitivty and Outcomes**

Exploring the correlation between VLE activity, student outcomes, and the number of assessments per course, we can find interesting insights that help validate some modules' difficulty.

The first thing to note is the correlation matrix below will result higher than expected due to variables being dependent on each other. For example, withdraw will naturally be a strong negative correlation to passing+distinction since they are essentially the same but worded differently - these situations will not be included in the interpretation.

* There is a medium correlation (0.5) between students that **failed** and the total number of assessments per course. This implies that more assessments are causing students to struggle, and material may be difficult.
  * Similarly, there is a negative relationship between **pass** and **quiz**. More quizzes in a module result in fewer students passing the calls, evidently meaning the VLE is too tricky.
  * Courses with more **forumng** negatively impact students graduating with **distinction with a negative correlation of -0.6.
* There is a medium correlation (0.5) between students that **withdraw** and **VLE Count**, meaning that more VLE activities drive students to leave the course. More work = more effort.


````
query = """
with assessments as (
  SELECT
    code_module,
    code_presentation,
    COUNT(id_assessment) AS assessment_cnt
  FROM `datasets-292703.adobe.assessments`
  GROUP BY 1,2
  ORDER BY 1,2)
  
, students AS (
  SELECT 
    r.code_module,
    r.code_presentation,
    count(distinct r.id_student) as total_students,
    round((count(distinct case when final_result IN ('Pass', 'Distinction') then i.id_student end) / count(distinct r.id_student))*100,2) as overall_passed,
    round((count(distinct case when final_result = 'Pass' then i.id_student end) / count(distinct r.id_student))*100,2)as pass,
    round((count(distinct case when final_result = 'Fail' then i.id_student end) / count(distinct r.id_student))*100,2)as fail,
    round((count(distinct case when final_result = 'Withdrawn' then i.id_student end) / count(distinct r.id_student))*100,2)as withdraw,
    round((count(distinct case when final_result = 'Distinction' then i.id_student end) / count(distinct r.id_student))*100,2)as distinction,
  FROM `datasets-292703.adobe.studentRegistration` as r
  JOIN `datasets-292703.adobe.studentInfo` as i
    on r.code_module = i.code_module and r.code_presentation = i.code_presentation and r.id_student = i.id_student
  group by 1,2
  order by 1,2)
  
, vles AS (
  SELECT
    code_module,
    code_presentation,
    COUNT(*) AS vle_cnt,
    COUNT(case when activity_type = 'forumng' then id_site end) as forumng,
    COUNT(case when activity_type = 'oucontent' then id_site end) as oucontent,
    COUNT(case when activity_type = 'subpage' then id_site end) as subpage,
    COUNT(case when activity_type = 'homepage' then id_site end) as homepage,
    COUNT(case when activity_type = 'quiz' then id_site end) as quiz,
  FROM
    `datasets-292703.adobe.vle`
  GROUP BY 1,2
  ORDER BY 1,2)
  
SELECT
  a.*,
  total_students,
  overall_passed as passing_and_distinction,
  pass,
  fail,
  withdraw,
  distinction,
  vle_cnt,
  forumng,
  oucontent,
  subpage,
  quiz
FROM assessments as a
JOIN students as s
  on a.code_module = s.code_module and a.code_presentation = s.code_presentation
JOIN vles as v
  on a.code_module = v.code_module and a.code_presentation = v.code_presentation
ORDER BY 1,2
"""

courses = client.query(query).to_dataframe()

courses.corr().round(1)
````
